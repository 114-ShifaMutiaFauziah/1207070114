```python
# Mengimport Library yang diperlukan

import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
# Membaca Gambar pada file menggunakan addresss dan disimpan dalam variabel 'img' 

img = imageio.imread("C:/Users/Asus/Downloads/PRAKTIKUM 6/Taj.jpg")
```

```python
# Mendapatkan resolusi dan type dari gambar

img_height = img.shape[0] # kodingan untuk menyimpan tinggi gambar
img_width = img.shape[1] # kodingan untuk menyimpan lebar gambar 
img_channel = img.shape[2] # kodingan untuk menyimpan jumlah saluran gambar 
img_type = img.dtype # kodingan untuk menyimpan tipe data piksel dalam gambar 
```

```python
# Membuat variabel dengan resolusi dan tipe yang sama seperti gambar
# kodingan ini untuk membuat array kosong yg akan digunakan untuk hasil gambar
# Array yang dibuat memiliki ukuran dan tipe data yang sama dengan gambar asli, yang diambil dari img.shape dan img_type

img_flip_horizontal = np.zeros(img.shape, img_type) 
img_flip_vertical = np.zeros(img.shape, img_type)
```

```python
# Kodingan untuk membalik gambar secara horizontal

for y in range(0, img_height):
    for x in range(0, img_width):
        for c in range(0, img_channel):
            img_flip_horizontal[y][x][c] = img[y][img_width-1-x][c]
```

```python
# kodingan untuk membalik gambar secara vertical
for y in range(0, img_height):
    for x in range(0, img_width):
        for c in range(0, img_channel):
            img_flip_vertical[y][x][c] = img[img_height-1-y][x][c]
```

```python
# Menampilkan hasil gambar flip horizontal dan flip vertical

plt.imshow(img_flip_horizontal) 
plt.title("Flip Horizontal")
plt.show()
plt.imshow(img_flip_vertical)
plt.title("Flip Vertical")
plt.show()
```
